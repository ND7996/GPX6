```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Directory containing the CSV file
directory_path = "/home/hp/results/mousecys/WT/"
csv_filename = "filtered_single_values_no_missing_or_outliers.csv"
csv_path = os.path.join(directory_path, csv_filename)

# Load data from CSV
df = pd.read_csv(csv_path)

# Check the original DataFrame
print("Original DataFrame:")
print(df.head())  # Display the first few rows of the DataFrame

# Check if required columns are present
required_columns = {'Mean dG*', 'StdError dG*', 'Mean dG0', 'StdError dG0'}
if required_columns.issubset(df.columns):
    # Log how many rows are present
    print(f"\nRows in the DataFrame: {len(df)}")

    # Extract values for plotting, ignoring errors for now
    all_dg_star_values = df['Mean dG*'].tolist()
    all_dg0_values = df['Mean dG0'].tolist()

    # Plotting
    plt.figure(figsize=(40, 20))

    # Scatter plot for Mean dG*
    plt.plot(range(1, len(all_dg_star_values) + 1), all_dg_star_values, 
             'o-', label='Mean dG*', markersize=8)

    # Scatter plot for Mean dG0
    plt.plot(range(1, len(all_dg0_values) + 1), all_dg0_values, 
             'o-', label='Mean dG0', markersize=8)

    # Customizing the plot
    plt.xlabel('Replica Index')
    plt.ylabel('Free Energy (kcal/mol)')
    plt.title('Free Energy Changes for Mouse Cys')
    plt.xticks(range(1, len(all_dg_star_values) + 1))  # Set x-ticks according to number of entries
    plt.legend()
    plt.grid()
    plt.show()
else:
    print("CSV does not contain the required columns.")
```
