```python
import matplotlib.pyplot as plt
import numpy as np

#sec.N        sec.CA       sec.CB       sec.SE              0.2   3.0        0.0
# Define the torsion parameters
V1 = 0.2
V2 = 3.0
V3 = 180.
V4 = 1.
gamma1 = 0.
gamma2 = 180.0
n = 1

# Define the torsion angle range
theta = np.linspace(-np.pi, np.pi, 361)

# Calculate the torsion potential energy for each angle
V =0.5*( V1*(1 + np.cos(n*(theta - gamma1))) + V2*(1 + np.cos(2*n*theta - gamma1)) + V3*(1 + np.cos(3*n*theta - gamma1)) + V4*(1 + np.cos(4*n*theta - gamma1)))

# Plot the torsion potential energy as a function of torsion angle
plt.plot(theta, V)
plt.xlabel('Torsion angle (radiants)')
plt.ylabel('Potential energy (kcal/mol)')
plt.title('OPLS torsion potential')
plt.show()
```
