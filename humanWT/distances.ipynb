```python
import os
import matplotlib.pyplot as plt

def read_dat_file(data_file):
    """Read data from a .dat file."""
    x_values, y_values = [], []
    with open(data_file, 'r') as file:
        for line in file:
            # Split the line into values
            values = line.strip().split()
            # Check if the line contains at least two values
            if len(values) >= 2:
                # Extract x and y values
                x, y = map(float, values[:2])
                x_values.append(x)
                y_values.append(y)
    return x_values, y_values

# Directory containing .dat files
directory = "/home/nayanika/github/PhD_Thesis/EVB/protein_stepwise/GPX6WT/mousecys/1-prep"  # Change this to your directory path

# Create a larger plot
plt.figure(figsize=(10, 8))

# Plot each .dat file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".dat"):
        data_file = os.path.join(directory, filename)
        # Read data from .dat file
        x_values, y_values = read_dat_file(data_file)
        # Plot the data
        plt.plot(x_values, y_values, label=filename[:-4])  # Use filename as label
        plt.xlabel('Reaction Coordinates/Frames', fontsize=14)
        plt.ylabel('Distances in Å', fontsize=14)
        plt.title('Human Selenocysteine', fontsize=16)
        plt.legend(fontsize=12)
        plt.grid(True)

plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```
